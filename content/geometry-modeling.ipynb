# Skinning: Surface Interpolation of curves#

"Skinning" is a basic incredient of CAD modelling. It is used to connect curves to surfaces.


# Gordon Surfaces: Interpolation of curve networks  #

The algorithm for curve network interpolation generates a surface from a network of profile curves and guide curves:


This is achieved by creating a superposition of three interpolation surfaces:


$$
\begin{align}
\color{gray}{s(u,v)} &= \color{green}{\sum^N_{i=1} f_i(u) \psi_i(v)} + \color{blue}{\sum^M_{j=1} g_j(v) \phi_j(u)} - \color{red}{\sum^N_{i=1}\sum^M_{j=1} \vec  \alpha_{i,j} \phi_j(u)\psi_i(v)} \notag \\
&= \color{green}{S_f(u,v)} + \color{blue}{S_g(u,v)} - \color{red}{T(u,v)}
\end{align}
$$
